# Benchmaq Python API

Run benchmarks using the Python API with config YAML files.

## Setup

Export your RunPod API key before running:
```bash
export RUNPOD_API_KEY="rpa_xxxxx"
```

````python
import os
import benchmaq

# Set Runpod API key
os.environ["RUNPOD_API_KEY"] = ""
````

````python
## run local gpu server
## no need to include `remote` section in config file
# result = benchmaq.bench("1_run_single.yaml")

## ssh remote 
## include server details in `remote` section in config file
result = benchmaq.bench("5_config_remote_ssh.yaml")
````

````python
## runpod remote
## include runpod details in `runpod` section in config file
result = benchmaq.runpod.bench("5_config_runpod.yaml")
````

````python
from multiprocessing import Pool
from benchmaq.runpod import run_benchmark  # Multiprocessing-compatible function

configs = [
    "5_config_runpod_multiprocess_1.yaml",  # 1x GPU
    "5_config_runpod_multiprocess_2.yaml",  # 2x GPU
]

with Pool(processes=len(configs)) as pool:
    results = pool.map(run_benchmark, configs)
````
